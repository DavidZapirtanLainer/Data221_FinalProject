```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Why: We manually input the final metrics from each individual notebook.
# Why: This allows us to visualize the comparison without having to re-run the training pipelines.
results = [
    {'Model': 'KNN', 'Accuracy': 0.9130, 'Precision': 0.9135, 'Recall': 0.9216, 'F1-Score': 0.9175, 'ROC-AUC': 0.9416},
    {'Model': 'Neural Network', 'Accuracy': 0.8913, 'Precision': 0.9020, 'Recall': 0.9020, 'F1-Score': 0.9020, 'ROC-AUC': 0.9260},
    {'Model': 'SVM', 'Accuracy': 0.8750, 'Precision': 0.8559, 'Recall': 0.9314, 'F1-Score': 0.8920, 'ROC-AUC': 0.9304},
    {'Model': 'Logistic Regression', 'Accuracy': 0.8859, 'Precision': 0.8716, 'Recall': 0.9314, 'F1-Score': 0.9005, 'ROC-AUC': 0.9293},
    {'Model': 'Random Forest', 'Accuracy': 0.9022, 'Precision': 0.9038, 'Recall': 0.9216, 'F1-Score': 0.9126, 'ROC-AUC': 0.9424},
    {'Model': 'Decision Tree', 'Accuracy': 0.8478, 'Precision': 0.8776, 'Recall': 0.8431, 'F1-Score': 0.8600, 'ROC-AUC': 0.8842}
]

# Why: Converting the list to a DataFrame allows us to easily access columns for plotting.
df_results = pd.DataFrame(results)

# Why: We define the standard classification metrics we want to analyze.
metrics = ["Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC"]

# Why: We define 6 specific colors to match our 6 models.
# Why: By keeping the order fixed, KNN will always be Blue, Neural Network Orange, etc.
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

for metric in metrics:
    # Why: We removed the sort_values() line so the X-axis order and color mapping remain identical across all plots.
    plt.figure(figsize=(10, 6))

    # Why: Using the fixed DataFrame ensures that the index-to-color mapping is consistent.
    plt.bar(df_results['Model'], df_results[metric], color=colors)

    # Why: We set the Y-axis range from 0.80 to 1.0 to focus on the small differences between high-performing models.
    plt.ylim(.80, 1.0)
    plt.ylabel('Score')
    plt.title(f'Model Performance Comparison: {metric}')
    plt.xticks(rotation=45)

    # Why: Adding the exact numeric value above each bar allows for precise comparison.
    for i, value in enumerate(df_results[metric]):
        # Why: The value + 0.005 offset ensures the text sits just above the bar without overlapping.
        plt.text(i, value + 0.005, f'{value:.4f}', ha='center', fontweight='bold')

    plt.tight_layout()
    plt.show()
```
