```python
#Data Cleaning and Sorting
import pandas as pd

# We need to encode the following Columns from Strings to a numerical value:
# Sex
# ExerciseAngina
# ChestPainType
# RestingECG
# ST_Slope


# Why: Loading the raw dataset into memory.
heart_data = pd.read_csv("heart.csv")

# Why: Binary features (Sex/Angina) are mapped to 0/1 integers to save memory
# and make them compatible with machine learning models.
# Why: Sex_F is used so the '1' explicitly represents Female.
heart_data['Sex_F'] = heart_data['Sex'].map({'M': 0, 'F': 1})
heart_data['ExerciseAngina'] = heart_data['ExerciseAngina'].map({'N': 0, 'Y': 1})

# Why: We remove the original string 'Sex' column now that we have 'Sex_F'.
heart_data = heart_data.drop(columns=['Sex'])

# Why: We define custom orders for our categories to control which one is "dropped."
# Why: The first item in each list will be the "baseline" (0,0,0) for the model.
# Why: ASY is the baseline for ChestPain, Normal is the baseline for ECG, and Up is the baseline for Slope.
heart_data['ChestPainType'] = pd.Categorical(heart_data['ChestPainType'], categories=['ASY', 'ATA', 'NAP', 'TA'])
heart_data['RestingECG'] = pd.Categorical(heart_data['RestingECG'], categories=['Normal', 'LVH', 'ST'])
heart_data['ST_Slope'] = pd.Categorical(heart_data['ST_Slope'], categories=['Up', 'Flat', 'Down'])

# Why: By calling get_dummies on the entire list at once, we ensure consistent column naming.
# Why: drop_first=True will now automatically drop our medically "Normal" or "Baseline" categories defined above.
categorical_cols = ['ChestPainType', 'RestingECG', 'ST_Slope']
heart_data = pd.get_dummies(heart_data, columns=categorical_cols, drop_first=True, dtype=int)

# Why: Displaying the info to verify that all columns are now numerical (int or float).
print(heart_data.info())
print(heart_data.head())
```
